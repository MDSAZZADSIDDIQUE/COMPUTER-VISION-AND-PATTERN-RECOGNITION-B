<h1>What is Activation Function?</h1>
<p>An activation function is a mathematical function used in artificial neural networks to introduce nonlinearity in the output of a neuron. The activation function takes the weighted sum of the input values to the neuron and applies a non-linear transformation to it, producing an output value that is then passed on to the next layer of the neural network.

The purpose of the activation function is to introduce nonlinearity into the neural network, allowing it to model complex relationships between the inputs and the output. Without nonlinearity, the neural network would only be able to model linear relationships between the inputs and the output, which limits its capabilities.</p>

<h1 style="padding: 2%; text-align: center;">Step Function</h1>
<p>The step function is a type of activation function used in artificial neural networks. It takes an input value and returns a binary output, either 0 or 1, depending on whether the input is less than or greater than or equal to a certain threshold value.</p>
<p><h3>Equation:</h3></p>

<h3>Nature:</h3>
<p>The nature of the step activation function is discontinuous, meaning that it has abrupt changes in its output when the input crosses a threshold.</p>
<h3>Value Range:</h3>
<p>0 or 1</p>
<h3>Uses:</h3>
<p>The step activation function is a simple but useful function that finds its applications in various fields. Here are some of the common uses of the step activation function:</p>
<ol><li><b>Binary Classification:</b> The step activation function is particularly useful in binary classification problems where the goal is to classify an input into one of two categories. In such cases, the output of the neuron is either 0 or 1, indicating whether the input belongs to one category or the other.</li>
<li><b>Perceptron Algorithm:</b> The step activation function is used in the perceptron algorithm, a simple linear binary classification algorithm used in machine learning. The perceptron algorithm learns a decision boundary to separate the data into two classes using the step activation function.</li>
<li><b>Perceptron Algorithm:</b> The step activation function is used in the perceptron algorithm, a simple linear binary classification algorithm used in machine learning. The perceptron algorithm learns a decision boundary to separate the data into two classes using the step activation function.</li>
<li><b>Perceptron Algorithm:</b> The step activation function is used in the perceptron algorithm, a simple linear binary classification algorithm used in machine learning. The perceptron algorithm learns a decision boundary to separate the data into two classes using the step activation function.</li>
</ol>
<p>While the step activation function has some limitations due to its discontinuity, it remains a useful function in certain applications where binary classification is required.</p>
<h3>Code:</h3>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def step_activation(x):
    return np.where(x >= 0, 1, 0)
```

```python
x = np.linspace(-10, 10, 100)
y = step_activation(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("step_activation(x)")
plt.title("Step activation function")
plt.show()
```

<h1>Sigmoid Function</h1>
<p>The sigmoid function is a mathematical function commonly used as an activation function in artificial neural networks. It is a type of logistic function that maps any input value to an output value between 0 and 1.</p>
<h3>Equation</h3>

<h3>Nature:</h3>
<p>The nature of the sigmoid function can be described as follows:</p>
<ol>
 <li><b>Nonlinear:</b> The sigmoid function is a nonlinear function, which means that the output does not vary linearly with the input. This nonlinearity is important in neural networks because it allows them to model complex relationships between the inputs and the output.</li>
<li><b>Smooth:</b> The sigmoid function is a smooth function, which means that it has continuous derivatives. This property makes it easy to optimize the function using gradient-based methods.</li>
<li><b>Saturation:</b> The sigmoid function can suffer from saturation, which means that the output of the function becomes insensitive to changes in the input when the input is too large or too small. This can make it difficult to optimize the function using gradient-based methods, especially when dealing with deep neural networks.</li>
<li><b>Probabilistic:</b> The sigmoid function can be interpreted as a probability distribution function, where the output value represents the probability of a particular input belonging to a certain class. This property makes the sigmoid function useful in binary classification problems.</li>
</ol>
<p>Overall, the sigmoid function has some limitations due to saturation, but it remains a useful function in certain applications, especially in binary classification problems where the output needs to be interpreted as a probability.</p>
<h3>Value Range:</h3>
<p>0 to 1</p>
<h3>Uses:</h3>
<ol>
<li><b>Binary Classification:</b> The sigmoid function is useful in binary classification problems, where the output needs to be interpreted as the probability of a particular input belonging to a particular class. The sigmoid function can produce a value between 0 and 1, which can be interpreted as the probability of the input belonging to the positive class.</li>
<li><b>Binary Classification:</b> The sigmoid function is useful in binary classification problems, where the output needs to be interpreted as the probability of a particular input belonging to a particular class. The sigmoid function can produce a value between 0 and 1, which can be interpreted as the probability of the input belonging to the positive class.</li>
<li><b>Logistic Regression:</b> The sigmoid function is also used in logistic regression, where it is used to model the probability of an event occurring.</li>
<li><b>Data Normalization:</b> The sigmoid function can also be used for data normalization. The sigmoid function can map any input value to a value between 0 and 1, which can help to normalize the input data.</li>
</ol>
<p>Overall, the sigmoid function has various uses in machine learning, such as in binary classification, artificial neural networks, logistic regression, and data normalization. However, the sigmoid function can suffer from saturation, which can make it difficult to optimize using gradient-based methods. As a result, other activation functions, such as the rectified linear unit (ReLU) function, have become more popular in recent years.</p>
<h3>Code:</h3>

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100)
y = sigmoid(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.title("Sigmoid activation function")
plt.show()
```

<h1>Relu</h1>
<p>ReLU stands for Rectified Linear Unit, which is an activation function used in artificial neural networks.</p>
<h3>Equation:</h3>

f(x) = max(0, x)

<h3>Nature:</h3>
<p>The nature of the ReLU function is as follows:</p>
<ol>
<li><b>Nonlinear:</b> The ReLU function is a nonlinear function because its output is not proportional to its input. When the input is negative, the output is zero, and when the input is positive, the output is equal to the input. This nonlinearity is essential for neural networks to model complex relationships between the inputs and the output.</li>
<li><b>Sparsity-inducing:</b> The ReLU function is sparsity-inducing because it outputs zero for any negative input value. This sparsity helps in reducing the number of connections between neurons and can make the network more computationally efficient.</li>
<li><b>Simple:</b> The ReLU function is a simple function that is easy to implement and computationally efficient. Its simplicity makes it well-suited for large-scale neural networks.</li>
<li><b>Zero-centered:</b> The ReLU function is not zero-centered, which means that the mean of the output is positive. This can be problematic when using the ReLU function as an activation function in certain types of neural networks.</li>
<li><b>Dying ReLU problem:</b> The ReLU function can suffer from a problem called the "dying ReLU problem" where some neurons in the network may become inactive and output zero for any input value. This can cause the network to lose its ability to learn and can lead to poor performance.</li>
</ol>
<p>In summary, the ReLU function is a nonlinear, sparsity-inducing, and simple activation function that is commonly used in artificial neural networks. Its properties make it suitable for large-scale networks, but it can suffer from the dying ReLU problem if not used carefully.</p>
<h3>Value Range:</h3>
<p>0 to infinity</p>
<h3>Uses:</h3>
<p>The Rectified Linear Unit (ReLU) function is one of the most commonly used activation functions in artificial neural networks. Some of the key uses of the ReLU function are:</p>
<ol>
<li><b>Efficient training:</b> The ReLU function is computationally efficient, making it well-suited for large-scale neural networks. It is also easy to implement and its derivatives can be calculated analytically, which makes it useful for gradient-based optimization methods like backpropagation.</li>
<li><b>Efficient training:</b> The ReLU function is computationally efficient, making it well-suited for large-scale neural networks. It is also easy to implement and its derivatives can be calculated analytically, which makes it useful for gradient-based optimization methods like backpropagation.</li>
<li><b>Sparsity:</b> The ReLU function outputs zero for any negative input value, which can help in reducing the number of connections between neurons and making the network more computationally efficient. This sparsity can also help in preventing overfitting, which occurs when the network becomes too complex and starts to fit the noise in the training data.</li>
<li><b>Sparsity:</b> The ReLU function outputs zero for any negative input value, which can help in reducing the number of connections between neurons and making the network more computationally efficient. This sparsity can also help in preventing overfitting, which occurs when the network becomes too complex and starts to fit the noise in the training data.</li>
</ol>
<p>It's worth noting that the ReLU function is not always the best choice for every neural network architecture or application. For example, it can suffer from the "dying ReLU" problem, where some neurons become inactive and output zero for any input value, leading to poor performance. In such cases, other variants of the ReLU function, such as the leaky ReLU or the parametric ReLU, may be more appropriate.</p>
<h3>Code:</h3>

```python
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.linspace(-10, 10, 100)
y = relu(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("relu(x)")
plt.title("Relu activation function")
plt.show()
```

<h1>Elu</h1>
<p>The Exponential Linear Unit (ELU) function is a type of activation function used in artificial neural networks.</p>
<h3>Equation:</h3>

<h3>Nature:</h3>
<p>The nature of the ELU function is such that it is smooth and continuous, with a smooth gradient. This makes it easier to optimize than some other activation functions that have sharp edges or are not differentiable at some points. Additionally, the ELU function can help to mitigate the "dying ReLU" problem that can occur with the ReLU function.</p>
<h3>Value Range:</h3>
<p>Infinity to -Infinity</p>
<h3>Uses:</h3>
<p>Some potential benefits of using the ELU function in neural networks include:</p>
<ol>
    <li><b>Improved performance:</b> The ELU function has been shown to perform well on a range of tasks, including image classification, object detection, and natural language processing.</li>
    <li><b>Reduced overfitting:</b> The ELU function can help to reduce overfitting by allowing neurons to take on negative values, which can help to regularize the network and prevent it from fitting too closely to the training data.</li>
    <li><b>Smoothness:</b> The ELU function is smooth and differentiable everywhere, which can help to speed up the optimization process and prevent numerical instability.</li>
</ol>
<p>Overall, the ELU function is a useful alternative to the ReLU function in many cases, and is worth considering when designing and training neural networks.</p>
<h3>Code</h3>

```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
x = np.linspace(-10, 10, 100)
y = elu(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("elu(x)")
plt.title("Elu activation function")
plt.show()
```

<h1>Selu</h1>
<p>The Scaled Exponential Linear Unit (SELU) function is an activation function used in artificial neural networks</p>
<h3>Equation:</h3>

<h3>Nature:</h3>
<p>The nature of the SELU function is such that it is self-normalizing. This means that if the weights of the network are initialized properly, the output of each layer will have mean zero and unit variance. This can help to speed up the training process and improve the performance of the network.</p>
<h3>Value Range:</h3>
<p>Infinity to -Infinity</p>
<h3>Uses:</h3>
<p></p>
<ol>
<li><b>Self-normalization:</b> The SELU function is self-normalizing, which means that it can help to speed up the training process and improve the performance of the network.</li>
<li><b>Improved performance:</b> The SELU function has been shown to perform well on a range of tasks, including image classification, object detection, and natural language processing.</li>
<li><b>Automatic adjustment of the activation function:</b> The SELU function can automatically adjust its parameters based on the input data, which can help to improve the performance of the network in different contexts.</li>
</ol>
<p>Overall, the SELU function is a useful alternative to other activation functions in many cases, and is worth considering when designing and training neural networks. However, it is important to note that the self-normalization property of the SELU function only holds when certain conditions are met, such as the weights of the network being initialized properly.</p>
<h3>Code:</h3>

```python
def selu(x, alpha=1.67326, scale=1.0507):
    return scale * np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
x = np.linspace(-10, 10, 100)
y = selu(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("selu(x)")
plt.title("Selu activation function")
plt.show()
```
